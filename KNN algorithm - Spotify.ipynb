```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
```

```python
features=["danceability","energy","tempo",'loudness','popularity','loudness','speechiness','speechiness']
s=StandardScaler()
```

```python
f_scaled=s.fit_transform(p[features])
f_scaled
```

```python
model=NearestNeighbors(n_neighbors=3,metric="euclidean")
model.fit(f_scaled)
```

```python
si=p[p["track_name"]=="To Begin Again"].index[0]
distances,indices=model.kneighbors([f_scaled[si]])
print("songs similar",p["track_name"])
for i in indices[0][1:]:
    print(p["track_name"][i])
```
